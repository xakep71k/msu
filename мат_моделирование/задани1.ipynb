```python
import math 

def systemSolution(beta0, V0x, V0z, m, t):
    x0 = 0
    z0 = 0
    Vx0 = 1
    Vz0 = V0z / V0x
    beta = beta0 * 2 * V0z / m * 9.8
    if beta == 0:
        Vx = Vx0
        Vz = Vz0 * (1 - 2 * t)
        x = x0 + Vx0 * t
        z = x0 + Vz0 * (t - t ** 2)
    else:
        Vx = Vx0 * math.exp(-beta * t)
        Vz = Vz0 * math.exp(-beta * t) * (1 + 2 / beta) - 2 * Vz0 / beta
        x = x0 + Vx0 / beta * (1 - math.exp(-beta * t))
        z = z0 + Vz0 / beta * (1 + 2 / beta) * (1 - math.exp(-beta * t)) - 2 * Vz0 / beta * t
        
    return (Vx, Vz, x, z)
```

```python
import numpy as np

beta0 = 0
V0x = 1
V0z = 1
m = 1

x1 = np.zeros(101)
z1 = np.zeros(101)
Vx1 = np.zeros(101)
Vz1 = np.zeros(101)

for i in range(101):
    (Vx1[i], Vz1[i], x1[i], z1[i]) = systemSolution(beta0, V0x, V0z, m, i/100.0)
```

```python
import matplotlib.pyplot as plt

t1 = []
for i in range(101):
    t1.append(i / 100.0)
fig, ax = plt.subplots()
ax.plot(t1, x1, linewidth = 3, color = 'g', label = 'x(t)')
ax.plot(t1, z1, linewidth = 3, color = 'y', label = 'z(t)')
ax.legend()
fig.set_figwidth(8)
fig.set_figheight(6)
plt.xlabel('t')
plt.ylabel('f(t)')
```

```python
figx, ax = plt.subplots()
ax.plot(t1, Vx1, linewidth = 3, color = 'g', label = 'Vx')
ax.plot(t1, Vz1, linewidth = 3, color = 'y', label = 'Vz')
ax.legend()
fig.set_figwidth(8)
fig.set_figheight(6)
plt.xlabel('t')
plt.ylabel('f(t)')
```

```python
fig, ax = plt.subplots()
ax.plot(x1, z1, linewidth = 3, color = 'm', label = 'z(x)')
ax.legend()
fig.set_figwidth(8)
fig.set_figheight(6)
plt.xlabel('x')
plt.ylabel('y')

#    
```
